```python
def model(train_data,y_train,X_test,y_test):
  # Dropping the duplicate values
  preprocessed_train =  pd.DataFrame()
  preprocessed_test =  pd.DataFrame()
  train_data = train_data.drop_duplicates()
  X_test = X_test.drop_duplicates()
  # # Replacing the missing values with appropriate statistic
  train_data['FinelineNumber'] = train_data['FinelineNumber'].fillna(train_data['FinelineNumber'].mode()[0], inplace = True)
  
  X_test['FinelineNumber'] = X_test['FinelineNumber'].fillna(X_test['FinelineNumber'].mode()[0], inplace = True)
  
  train_data['Upc'] = train_data['Upc'].fillna(train_data['Upc'].mode()[0], inplace = True)
  
  X_test['Upc'] = X_test['Upc'].fillna(X_test['Upc'].mode()[0], inplace = True)
  # train_data['DepartmentDescription'] = train_data['DepartmentDescription'].fillna('OTHER DEPARTMENTS', inplace = True)


  # Here in DepartmentDescription we are combining the categories of similar type, this will make the categories with less count fall under one major category. 

  train_data["DepartmentDescription"] = train_data["DepartmentDescription"].apply(lambda x: str(x).replace("MENSWEAR", "MENS WEAR"))
  train_data["DepartmentDescription"] = train_data["DepartmentDescription"].apply(lambda x: str(x).replace("LADIES SOCKS", "LADIESWEAR"))
  train_data["DepartmentDescription"] = train_data["DepartmentDescription"].apply(lambda x: str(x).replace("PLUS AND MATERNITY", "LADIESWEAR"))
  train_data["DepartmentDescription"] = train_data["DepartmentDescription"].apply(lambda x: str(x).replace("BRAS & SHAPEWEAR", "LADIESWEAR"))
  train_data["DepartmentDescription"] = train_data["DepartmentDescription"].apply(lambda x: str(x).replace("PHARMACY RX", "PHARMACY"))
  train_data["DepartmentDescription"] = train_data["DepartmentDescription"].apply(lambda x: str(x).replace("PHARMACY OTC", "PHARMACY"))
  train_data["DepartmentDescription"] = train_data["DepartmentDescription"].apply(lambda x: str(x).replace("HEALTH AND BEAUTY AIDS", "PHARMACY"))
  train_data["DepartmentDescription"] = train_data["DepartmentDescription"].apply(lambda x: str(x).replace("INFANT CONSUMABLE HARD LINES", "INFANT SECTION"))
  train_data["DepartmentDescription"] = train_data["DepartmentDescription"].apply(lambda x: str(x).replace("INFANT APPAREL", "INFANT SECTION"))
  train_data["DepartmentDescription"] = train_data["DepartmentDescription"].apply(lambda x: str(x).replace("OPTICAL - FRAMES", "OPTICALS"))
  train_data["DepartmentDescription"] = train_data["DepartmentDescription"].apply(lambda x: str(x).replace("OPTICAL - LENSES", "OPTICALS"))
  
  X_test["DepartmentDescription"] = X_test["DepartmentDescription"].apply(lambda x: str(x).replace("MENSWEAR", "MENS WEAR"))
  X_test["DepartmentDescription"] = X_test["DepartmentDescription"].apply(lambda x: str(x).replace("LADIES SOCKS", "LADIESWEAR"))
  X_test["DepartmentDescription"] = X_test["DepartmentDescription"].apply(lambda x: str(x).replace("PLUS AND MATERNITY", "LADIESWEAR"))
  X_test["DepartmentDescription"] = X_test["DepartmentDescription"].apply(lambda x: str(x).replace("BRAS & SHAPEWEAR", "LADIESWEAR"))
  X_test["DepartmentDescription"] = X_test["DepartmentDescription"].apply(lambda x: str(x).replace("PHARMACY RX", "PHARMACY"))
  X_test["DepartmentDescription"] = X_test["DepartmentDescription"].apply(lambda x: str(x).replace("PHARMACY OTC", "PHARMACY"))
  X_test["DepartmentDescription"] = X_test["DepartmentDescription"].apply(lambda x: str(x).replace("HEALTH AND BEAUTY AIDS", "PHARMACY"))
  X_test["DepartmentDescription"] = X_test["DepartmentDescription"].apply(lambda x: str(x).replace("INFANT CONSUMABLE HARD LINES", "INFANT SECTION"))
  X_test["DepartmentDescription"] = X_test["DepartmentDescription"].apply(lambda x: str(x).replace("INFANT APPAREL", "INFANT SECTION"))
  X_test["DepartmentDescription"] = X_test["DepartmentDescription"].apply(lambda x: str(x).replace("OPTICAL - FRAMES", "OPTICALS"))
  X_test["DepartmentDescription"] = X_test["DepartmentDescription"].apply(lambda x: str(x).replace("OPTICAL - LENSES", "OPTICALS"))

# Now applying the pd.get_dummies on DepartmentDescription

  dept_feature = []
  for i in list(pd.get_dummies(train_data[['DepartmentDescription']]).columns):
    dept_feature.append(i)
  
  dept = pd.get_dummies(train_data[['DepartmentDescription']])
  train_data  = pd.concat([train_data, dept],axis = 1 )
# Iterating through dept description columns and grouping by based on Visit number
  for j in dept_feature:
    preprocessed_train[j] = train_data.groupby(['VisitNumber'])[j].sum()
    
  dept_feature = []
  for i in list(pd.get_dummies(X_test[['DepartmentDescription']]).columns):
    dept_feature.append(i)
  
  dept_test = pd.get_dummies(X_test[['DepartmentDescription']])
  X_test  = pd.concat([X_test, dept_test],axis = 1 )

# Iterating through dept description columns and grouping by based on Visit number
  for j in dept_feature:
    preprocessed_test[j] = X_test.groupby(['VisitNumber'])[j].sum()
    
  preprocessed_test.drop('DepartmentDescription_nan',axis=1,inplace=True)

  preprocessed_train.drop('DepartmentDescription_nan',axis=1,inplace=True)
  
  #Created a new column 'Return' which marks as 1 if scan count is negative
  train_data['Return'] = train_data['ScanCount']
  train_data.loc[train_data['ScanCount'] > 0, 'Return'] = 0
  train_data.loc[train_data['ScanCount'] < 0, 'Return'] = 1

  # Aggregating based on VisitNumber for all the columns 
  #https://www.tutorialspoint.com/python_pandas/python_pandas_groupby.htm
  Upc = train_data.groupby(['VisitNumber'])['Upc'].count()
  preprocessed_train['Upc'] = Upc
  FinelineNumber = train_data.groupby(['VisitNumber'])['FinelineNumber'].count()
  preprocessed_train['FinelineNumber'] = FinelineNumber
  Return = train_data.groupby(['VisitNumber'])['Return'].max()
  preprocessed_train['Return'] = Return
  Weekday = train_data.groupby(['VisitNumber'])['Weekday'].max()
  preprocessed_train['Weekday'] = Weekday
  trip = train_data
  trip['Trip_type'] = y_train
  TripType = trip.groupby(['VisitNumber'])['Trip_type'].max()
  preprocessed_train['TripType'] = TripType
  ScanCount = train_data.groupby(['VisitNumber'])['ScanCount'].max()
  preprocessed_train['ScanCount'] = ScanCount
  
  


  #Created a new column 'Return' which marks as 1 if scan count is negative
  X_test['Return'] = X_test['ScanCount']
  X_test.loc[X_test['ScanCount'] > 0, 'Return'] = 0
  X_test.loc[X_test['ScanCount'] < 0, 'Return'] = 1

  Upc = X_test.groupby(['VisitNumber'])['Upc'].count()
  preprocessed_test['Upc'] = Upc
  FinelineNumber = X_test.groupby(['VisitNumber'])['FinelineNumber'].count()
  preprocessed_test['FinelineNumber'] = FinelineNumber
  Return = X_test.groupby(['VisitNumber'])['Return'].max()
  preprocessed_test['Return'] = Return
  Weekday = X_test.groupby(['VisitNumber'])['Weekday'].max()
  preprocessed_test['Weekday'] = Weekday
  trip_test = X_test
  trip_test['Trip_type'] = y_test
  TripType = trip_test.groupby(['VisitNumber'])['Trip_type'].max()
  preprocessed_test['TripType'] = TripType
  ScanCount = X_test.groupby(['VisitNumber'])['ScanCount'].max()
  preprocessed_test['ScanCount'] = ScanCount
  
  preprocessed_train['Weekday'] = preprocessed_train['Weekday'].apply(lambda x: str(x).replace("Monday", '1'))
  preprocessed_train['Weekday'] = preprocessed_train['Weekday'].apply(lambda x:str(x).replace("Tuesday",'2'))
  preprocessed_train['Weekday'] = preprocessed_train['Weekday'].apply(lambda x:str(x).replace("Wednesday",'3'))
  preprocessed_train['Weekday'] = preprocessed_train['Weekday'].apply(lambda x:str(x).replace("Thursday",'4'))
  preprocessed_train['Weekday'] = preprocessed_train['Weekday'].apply(lambda x:str(x).replace("Friday",'5'))
  preprocessed_train['Weekday'] = preprocessed_train['Weekday'].apply(lambda x:str(x).replace("Saturday",'6'))
  preprocessed_train['Weekday'] = preprocessed_train['Weekday'].apply(lambda x:str(x).replace("Sunday",'7'))
  
  preprocessed_test['Weekday'] = preprocessed_test['Weekday'].apply(lambda x: str(x).replace("Monday", '1'))
  preprocessed_test['Weekday'] = preprocessed_test['Weekday'].apply(lambda x:str(x).replace("Tuesday",'2'))
  preprocessed_test['Weekday'] = preprocessed_test['Weekday'].apply(lambda x:str(x).replace("Wednesday",'3'))
  preprocessed_test['Weekday'] = preprocessed_test['Weekday'].apply(lambda x:str(x).replace("Thursday",'4'))
  preprocessed_test['Weekday'] = preprocessed_test['Weekday'].apply(lambda x:str(x).replace("Friday",'5'))
  preprocessed_test['Weekday'] = preprocessed_test['Weekday'].apply(lambda x:str(x).replace("Saturday",'6'))
  preprocessed_test['Weekday'] = preprocessed_test['Weekday'].apply(lambda x:str(x).replace("Sunday",'7'))


  #Random Forest
  y_train = preprocessed_train['TripType']
  preprocessed_train = preprocessed_train.drop(['TripType'],axis = 1)

  y_test = preprocessed_test['TripType']
  preprocessed_test = preprocessed_test.drop(['TripType'],axis = 1)
  
  rf = RandomForestClassifier(n_estimators=100)
  rf = rf.fit(np.asarray(preprocessed_train), np.asarray(y_train))

  

  return rf, preprocessed_test, preprocessed_train, y_train, y_test
```
